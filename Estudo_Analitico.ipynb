### Demonstra√ß√£o da quantidade de musica que s√£o lan√ßadas por ano, para identificarmos se estes anos est√£o influenciando no meu TOP 10 de musicas com mais streams.  

```python
year_counts = df['released_year'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
year_counts.plot(kind='bar', color='coral')
plt.title('N√∫mero de M√∫sicas Lan√ßadas por Ano')
plt.xlabel('Ano')
plt.ylabel('Contagem')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Supondo que 'year_counts' j√° foi criado a partir do DataFrame
year_counts = df['released_year'].value_counts().sort_index()

# Criar o gr√°fico de linha
plt.figure(figsize=(10, 6))
plt.plot(year_counts.index, year_counts.values, marker='o', linestyle='-', color='coral')

# Adicionar t√≠tulo e r√≥tulos
plt.title('N√∫mero de M√∫sicas Lan√ßadas por Ano')
plt.xlabel('Ano')
plt.ylabel('Contagem')

# Girar os r√≥tulos do eixo X para melhor legibilidade
plt.xticks(rotation=45)

# An√°lise: Anotar o ponto com maior n√∫mero de lan√ßamentos
max_year = year_counts.idxmax()  # Ano com o maior n√∫mero de lan√ßamentos
max_value = year_counts.max()    # Valor m√°ximo de lan√ßamentos

# Adicionar a anota√ß√£o ao gr√°fico
plt.annotate(f'M√°ximo em {max_year}', xy=(max_year, max_value),
             xytext=(max_year, max_value + 10),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Mostrar o gr√°fico
plt.show()

```
